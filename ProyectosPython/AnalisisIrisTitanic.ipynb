````
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head(5)
````

````
# Soluci√≥n
datos2 = titanic[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']] #Tabla con las variables a utilizar
datos2.head()
````

````
datos2.age = datos2.age.fillna(datos2.age.mean()) #imputacion de promedio de edad
nulos = datos2.isnull()
nulos.value_counts()
````

````
datos2 = datos2.dropna()
````

## <font color='green'>Actividad 4</font>

1. Entregar la informaci√≥n correspondiente a la la tasa de supervivencia promedio agrupada por sexo y clase del pasajero.
2. Genere una tabla que muestre la cantidad de personas agrupadas por sexo y clase.
3. Muestre la misma informaci√≥n del punto 1, pero la cantidad de sobrevivientes.
 


````
# Soluci√≥n
# Supervivencia promedio por sexo y clase
datos2.groupby(by = ['sex', 'pclass'])['survived'].mean() #Las mujeres de las clases superiores sobreviven mucho m√°s que los hombres, adem√°s los hombres de primera clase sobrevivieron m√°s que los hombres de clases inferiores
````

````
#cantidad de personas agrupadas por sexo y clase
datos2.groupby(by = ['sex', 'pclass'])['sex'].count() #La cantidad de mujeres era inferior a la de hombres en todas las clases
````

````
# Supervivencia neta por sexo y clase
vivieron = datos2.where(datos2['survived']==1)
vivieron.groupby(by = ['sex', 'pclass'])['survived'].count() #No solo la tasa de supervivencia femenina es mayor sino que adem√°s sobrevivieron m√°s mujeres que hombres en todas las clases
````

Hasta ahora hemos examinado las tasas de supervivencia seg√∫n el sexo y la clase de pasajeros. Otro factor que podr√≠a haber influido en la supervivencia es la edad. **¬øLos ni√±os varones ten√≠an tanta probabilidad de morir como las mujeres en general?** Podemos investigar esta pregunta mediante la indexaci√≥n m√∫ltiple, o pivotando en m√°s de dos variables, agregando otro √≠ndice.

Divida la informaci√≥n de la columna "age" en intervalos, creando as√≠ una nueva categor√≠a que se puede utilizar para agrupar. Divida el conjunto en intervalos de (0, 12], (12, 18] y (18, 80].

4. Genere la tabla que le ayude a responder la pregunta anterior (considerando los rangos definidos).
5. Para una mejor interpretaci√≥n de la informaci√≥n anterior, genere una tabla con el n√∫mero de entradas por cada grupo.






````
conditionlist = [
    (datos2['age'] > 18) ,
    (datos2['age'] > 12) & (datos2['age'] <=18),
    (datos2['age'] <= 12)]
choicelist = ['18-80', '12-18','0-12']
datos2['edad'] = np.select(conditionlist, choicelist, default='Not Specified')
````

````
# Soluci√≥n
tabla1 = datos2.pivot_table(index = 'edad', columns = 'sex', values = 'survived')  #N√±os y ni√±as de 0-12 tienen una supervivencia similar, en el resto de los grupos etarios las mujeres sobreviven mucho m√°s
tabla1
````

````
vivieron = datos2.where(datos2['survived']==1)
tabla2 = vivieron.pivot_table(index = 'edad', columns = 'sex', values = 'survived', aggfunc='count') #La cantidad de supervivientes es mayor en el rango de 18 a 80 tanto para hombres como para mujeres
tabla2 # pero esto se debe a que la mayor parte de los pasajeros correspond√≠an a este rango eterio
````

Pandas tambi√©n admite la indexaci√≥n m√∫ltiple en las columnas. Como ejemplo, considere el precio de los boletos de un pasajero. Esta es otra caracter√≠stica continua que se puede discretizar. Divida los precios en 2 cuantiles iguales (baratos y caros). Algunos de los grupos resultantes podr√≠an estar vac√≠os; para mejorar la legibilidad de la tabla use un gui√≥n en las casillas vac√≠as.

6. Genere una tabla que nos permita ver el efecto del precio de los tickets.

````
# Soluci√≥n
conditionlist = [
    (datos2['fare'] < datos2['fare'].quantile(0.5)) ,
    (datos2['fare'] > datos2['fare'].quantile(0.5))]
choicelist = ['Baratos', 'Caros']
datos2['tipoPrecio'] = np.select(conditionlist, choicelist, default='-')
````

````
tabla3 = datos2.pivot_table(index = 'tipoPrecio', columns = 'sex', values = 'survived') #Los valores en el - son losequivalentes a la mediana
tabla3 # La posibilidad de sobrevivir era levemente mayor tanto para hombres y mujeres si hab√≠an comprado un boleto m√°s caro
````
