````
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head(5)
````

````
# Solución
datos2 = titanic[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']] #Tabla con las variables a utilizar
datos2.head()
````

````
datos2.age = datos2.age.fillna(datos2.age.mean()) #imputacion de promedio de edad
nulos = datos2.isnull()
nulos.value_counts()
````

````
datos2 = datos2.dropna()
````

## <font color='green'>Actividad 4</font>

1. Entregar la información correspondiente a la la tasa de supervivencia promedio agrupada por sexo y clase del pasajero.
2. Genere una tabla que muestre la cantidad de personas agrupadas por sexo y clase.
3. Muestre la misma información del punto 1, pero la cantidad de sobrevivientes.
 


````
# Solución
# Supervivencia promedio por sexo y clase
datos2.groupby(by = ['sex', 'pclass'])['survived'].mean() #Las mujeres de las clases superiores sobreviven mucho más que los hombres, además los hombres de primera clase sobrevivieron más que los hombres de clases inferiores
````

````
#cantidad de personas agrupadas por sexo y clase
datos2.groupby(by = ['sex', 'pclass'])['sex'].count() #La cantidad de mujeres era inferior a la de hombres en todas las clases
````

````
# Supervivencia neta por sexo y clase
vivieron = datos2.where(datos2['survived']==1)
vivieron.groupby(by = ['sex', 'pclass'])['survived'].count() #No solo la tasa de supervivencia femenina es mayor sino que además sobrevivieron más mujeres que hombres en todas las clases
````

Hasta ahora hemos examinado las tasas de supervivencia según el sexo y la clase de pasajeros. Otro factor que podría haber influido en la supervivencia es la edad. **¿Los niños varones tenían tanta probabilidad de morir como las mujeres en general?** Podemos investigar esta pregunta mediante la indexación múltiple, o pivotando en más de dos variables, agregando otro índice.

Divida la información de la columna "age" en intervalos, creando así una nueva categoría que se puede utilizar para agrupar. Divida el conjunto en intervalos de (0, 12], (12, 18] y (18, 80].

4. Genere la tabla que le ayude a responder la pregunta anterior (considerando los rangos definidos).
5. Para una mejor interpretación de la información anterior, genere una tabla con el número de entradas por cada grupo.






````
conditionlist = [
    (datos2['age'] > 18) ,
    (datos2['age'] > 12) & (datos2['age'] <=18),
    (datos2['age'] <= 12)]
choicelist = ['18-80', '12-18','0-12']
datos2['edad'] = np.select(conditionlist, choicelist, default='Not Specified')
````

````
# Solución
tabla1 = datos2.pivot_table(index = 'edad', columns = 'sex', values = 'survived')  #Nños y niñas de 0-12 tienen una supervivencia similar, en el resto de los grupos etarios las mujeres sobreviven mucho más
tabla1
````

````
vivieron = datos2.where(datos2['survived']==1)
tabla2 = vivieron.pivot_table(index = 'edad', columns = 'sex', values = 'survived', aggfunc='count') #La cantidad de supervivientes es mayor en el rango de 18 a 80 tanto para hombres como para mujeres
tabla2 # pero esto se debe a que la mayor parte de los pasajeros correspondían a este rango eterio
````

Pandas también admite la indexación múltiple en las columnas. Como ejemplo, considere el precio de los boletos de un pasajero. Esta es otra característica continua que se puede discretizar. Divida los precios en 2 cuantiles iguales (baratos y caros). Algunos de los grupos resultantes podrían estar vacíos; para mejorar la legibilidad de la tabla use un guión en las casillas vacías.

6. Genere una tabla que nos permita ver el efecto del precio de los tickets.

````
# Solución
conditionlist = [
    (datos2['fare'] < datos2['fare'].quantile(0.5)) ,
    (datos2['fare'] > datos2['fare'].quantile(0.5))]
choicelist = ['Baratos', 'Caros']
datos2['tipoPrecio'] = np.select(conditionlist, choicelist, default='-')
````

````
tabla3 = datos2.pivot_table(index = 'tipoPrecio', columns = 'sex', values = 'survived') #Los valores en el - son losequivalentes a la mediana
tabla3 # La posibilidad de sobrevivir era levemente mayor tanto para hombres y mujeres si habían comprado un boleto más caro
````
