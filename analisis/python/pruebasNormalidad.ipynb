# Revisión de normalidad de datos - datos random

```python
import numpy as np
import matplotlib.pyplot as plt #graficos
data_points= np.random.normal(0,1,100) #datos generados aleatoreamente
```

```python
data_points
```

Revisión de histograma- grafico sigue forma de campana de gauss por ende nos indica que tiene distribución normal

```python
plt.hist(data_points,edgecolor='black', linewidth=1)
```

Revisión de grafico quantile quantile - si sigue la linea datos poseen distribución normal

```python
import pylab
import scipy.stats as stats
```

```python
stats.probplot(data_points,dist="norm", plot=pylab)
pylab.show()
```
