```python
dfangel = df.loc[df['Gastos'].str.contains('angel') & ~df['Gastos'].str.contains('kony') & ~df['Gastos'].str.contains('demi') & ~df['Gastos'].str.contains('debe')]
dfangel['valor'] = dfangel['Gastos'].str.extract('(\d+\.\d+|\d+)').astype(int)
dfangel
```

```python
totalangel = dfangel['valor'].sum()
totalangel
```

```python
dfkony = df.loc[df['Gastos'].str.contains('kony') & ~df['Gastos'].str.contains('angel') & ~df['Gastos'].str.contains('demi') & ~df['Gastos'].str.contains('debe')]
dfkony['valor'] = dfkony['Gastos'].str.extract('(\d+\.\d+|\d+)').astype(int)
dfkony
```

```python
totalkony = dfkony['valor'].sum()
totalkony
```

```python
dfdemi = df.loc[df['Gastos'].str.contains('demi') & ~df['Gastos'].str.contains('angel') & ~df['Gastos'].str.contains('kony') & ~df['Gastos'].str.contains('debe')]
dfdemi['valor'] = dfdemi['Gastos'].str.extract('(\d+\.\d+|\d+)').astype(int)
dfdemi
```

```python
totaldemi = dfdemi['valor'].sum()
totaldemi
```

```python
dfpagos = df.loc[df['Gastos'].str.contains('debe')]
dfpagos['valor'] = dfpagos['Gastos'].str.extract('(\d+\.\d+|\d+)').astype(int)
dfpagos
```

```python
print(f"Total Angel: {totalangel}")
print(f"Total Kony: {totalkony}")
print(f"Total Demi: {totaldemi}")

tercio_angel = totalangel / 3
tercio_kony = totalkony / 3
tercio_demi = totaldemi / 3

print(f"Tercio Angel: {tercio_angel}")
print(f"Tercio Kony: {tercio_kony}")
print(f"Tercio Demi: {tercio_demi}")
```
